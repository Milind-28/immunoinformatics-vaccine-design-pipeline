```python
from itertools import product, permutations

def generate_combinations(epitopes, linkers, linker_positions):

    # Generate all combinations of epitope variants
    epitope_combinations = list(product(*epitopes.values()))

    # Shuffle the epitope positions
    shuffled_combinations = []
    for epitope_combination in epitope_combinations:
        for permuted_epitopes in permutations(epitope_combination):
            # Insert linkers into the fixed positions
            for linker_combination in product(*linkers.values()):
                sequence = []
                linker_idx = 0
                for position in linker_positions:
                    if 'epitope' in position:
                        sequence.append(permuted_epitopes[int(position.split()[1]) - 1])
                    elif 'linker' in position:
                        sequence.append(linker_combination[linker_idx])
                        linker_idx += 1
                shuffled_combinations.append("".join(sequence))

    return shuffled_combinations

# Example usage
if __name__ == "__main__":
    # Define the multi-epitope sequence with variants for each position
    epitopes = {
        "epitope 1": ["_FARELAYEGLDPALIIKKLKETGGDDWV_"],
        "epitope 2": ["_DTKFIIVFALTRGNKIVKASGKMSNSGS_"],
        "epitope 3": ["_KEYLPVGPAVMNLKVENYPPEMMCMAFGSLIPT_"]
    }

    # Define linker variants
    linkers = {
        "linker 1": ["GGGGS"],
        "linker 2": ["GGGGS"],
    }

    # Define the sequence of positions (epitopes and linkers)
    linker_positions = ["epitope 1", "linker 1", "epitope 2", "linker 2", "epitope 3"]

    # Generate combinations
    combinations = generate_combinations(epitopes, linkers, linker_positions)

    # Print the results
    print("Total combinations:", len(combinations))
    for sequence in combinations:
        print(sequence)
```
